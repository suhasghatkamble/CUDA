```python
%%writefile numba_helloworld.py
import numpy as np
from numba import cuda

# CUDA kernel function
@cuda.jit
def hello_cuda():
  tx = cuda.threadIdx.x
  bx = cuda.blockIdx.x
  bw = cuda.blockDim.x

  idx = tx + bx * bw

  print("Hello, World! I'm thread", idx)

threads_per_block = 5
blocks_per_grid = 256 # Increase the number of blocks

hello_cuda[blocks_per_grid, threads_per_block]()   #decorators called
```
